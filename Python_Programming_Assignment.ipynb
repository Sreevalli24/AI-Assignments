```python
import pandas as pd
```

```python
df=pd.read_csv('/content/employee.csv')
df
```

Given an employee id, greet them with their first name.

```python
def greet_employee(eid):
    row = df[df['EID'] == eid]
    if not row.empty:
        name = row.iloc[0]['Employee Name']
        gender = row.iloc[0]['Gender']
        first_name = name.split()[0]
        prefix = "Mr" if gender.lower() == "male" else "Ms"
        print(f"Hello, {prefix} {first_name}")
    else:
        print("Employee ID not found.")
```

```python
greet_employee('E876G')
```

Function for calculating the average salary of each department.

```python
def average_salary_by_department():
    avg_salary = df.groupby('Department')['Salary'].mean()
    print(avg_salary)
average_salary_by_department()
```

Given an Employee id, display respective employees Name, Data of Joining and Department.

```python
def show_employee_details(eid):
    row = df[df['EID'] == eid]
    if not row.empty:
        name = row.iloc[0]['Employee Name']
        doj = row.iloc[0]['Date of Joining']
        dept = row.iloc[0]['Department']
        print(f"{name} | {doj} | {dept}")
    else:
        print("Employee ID not found.")
show_employee_details('E876G')
```

Function for sorting employees from Oldest to Newest.

```python
def sort_employees_by_joining():
    df['Date of Joining'] = pd.to_datetime(df['Date of Joining'], dayfirst=True)
    sorted_df = df.sort_values('Date of Joining')
    print(sorted_df[['EID', 'Employee Name', 'Date of Joining']])
sort_employees_by_joining()
```
