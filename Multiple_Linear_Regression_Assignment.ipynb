Profit Column has outliers

```python
#removing outliers
columns = ['R&D Spend', 'Administration', 'Marketing Spend', 'Profit']

for col in columns:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]
```

```python
plt.figure(figsize=(10,7))
sns.boxplot(data=df)
```

```python
x=df.drop('Profit', axis=1)
y=df['Profit']
```

```python
x
```

```python
y
```

```python
#onehot encoding
#Convert the column categorical columns

states=pd.get_dummies(x['State'],drop_first=True) # drop_first=True is used in pandas' get_dummies() to remove the first category when converting
#categorical variables into dummy/indicator variables. This helps avoid multicollinearity, especially in regression models,
#by eliminating redundant information while still preserving the complete meaning of the data.

# Drop the state coulmn
x=x.drop('State',axis=1) #axis=1 tells drop a column not a row, axis=0 means row

# concat the dummy variables
x=pd.concat([x,states],axis=1)
x
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
models=[]
```

```python
#Model1 : Linear Regression(OLS) - ordinary least square method
from sklearn.linear_model import LinearRegression
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
models.append(('OLS', LR))
```

```python
#Model2 : Lasso Regression
from sklearn.linear_model import Lasso
lasso=Lasso(alpha=0.1, random_state=42)
lasso.fit(x_train,y_train)
```

```python
models.append(('Lasso', lasso))
```

```python
#Model3: Ridge Regression
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=0.1, random_state=42)
ridge.fit(x_train,y_train)
```

```python
models.append(('Ridge', ridge))
```

**Evaluating Performance of Models**

```python
from sklearn.metrics import mean_squared_error, r2_score
for name, model in models:
    y_pred = model.predict(x_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f"{name} - Mean Squared Error: {mse}, R^2 Score: {r2}")
```

```python
#we infer it is a good model
```

**Applying Lasso and Ridge Methods**

```python
#Lasso Regression
lasso_model = Lasso(alpha=0.01)
lasso_model.fit(x_train, y_train)
lasso_coefficients = lasso_model.coef_
lasso_coefficients
```

```python
print("Lasso Regression Coefficients:")
for feature, coef in zip(x.columns, lasso_coefficients):
    print(f"{feature}: {coef}")
```

```python
#Ridge Regression
ridge_model = Ridge(alpha=1.0)
ridge_model.fit(x_train, y_train)
ridge_coefficients = ridge_model.coef_
ridge_coefficients
```

```python
print("Ridge Regression Coefficients:")
for feature, coef in zip(x.columns, ridge_coefficients):
    print(f"{feature}: {coef}")
```
